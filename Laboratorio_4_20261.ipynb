# **Pseudoalgoritmos**

Un pseudoalgoritmo es una forma generalizada de implementar un algoritmo sin necesidad de incluir la sintaxis de un lenguaje de programación en concreto. Se caracteriza por no ser realmente ejecutable sino una representación de la idea lógica que implica su funcionamiento para que pueda ser transmitida arbitrariamentre. 

Dado un pseudoalgoritmo, el programador es libre de llevar a cabo esta idea lógica en cualquier lenguaje de programación y ejecutarla de una manera personalizada. Un ejemplo de pseudoalgoritmo sería el siguiente:

Este pseudoalgoritmo generaliza la idea del famoso algoritmo *Insertion sort*, que organiza en orden ascendente los números contenidos en un arreglo. Para ello inicia describiendo la información de entrada y la que se obtendrá como resultado, es decir el arreglo desordenado y el arreglo ordenado respectivamente. Luego, muestra el proceso mediante el cual se opera el arreglo para ordenarse, usando estructuras iterativas.

En particular, podemos implementar este pseudoalgoritmo en Python de la siguiente forma:

```python
def InsertionSort(A:list):
  """
  Función que toma por argumento una lista de Python y retorna 
  nuevamente la lista con sus elementos en orden ascendente
  """
  N = len(A) 
  for j in range(2,N):
    key = A[j]
    i = j-1
    while (i>=0) and (A[i]>key):
      A[i+1]=A[i]
      i = i-1
    A[i+1] = key
  return A

A = [1, 9, 2, 7, 1, 6, 0, 4]
B = InsertionSort(A)
print(B)
```

Los pseudoalgoritmos son fundamentales para transmitir ideas algorítmicas sin necesidad de restringirlas a un lenguaje de programación en concreto. Estos son muy usados en textos de optimización, Machine Learning y Deep Learning debido a su versatilidad, por lo que la idea de este laboratorio es familiarizarse con ellos, aprender a leerlos y, posteriormente, ejecutarlos.

------------------------------
*Interpretación: El algoritmo planteado es uno que recibe dos enteros (a y b), y calcula el máximo común divisor. Para hacer esto, el algoritmo hace uso del Algoritmo de Euclides. Para esto, se revisa que A sea distinto a B. Mientras esto siga siendo el caso, el algoritmo repite el siguiente proceso. primero, se verifica que A es mayor a B, si es el caso, se le resta a A, b. Si A no es mayor que B, se le resta A a B. Posteriormente se revisa nuevamente que A sea distinto de B, si son distintos, se continua, de lo contrario, se termina.*
------------------------------ -

```python
#----------------------------------------------------
#Ejecución: implemente y ejecute su algoritmo aquí
#----------------------------------------------------

def EuclidianAlgorithm(a:int, b:int)-> int:
  """
  Función que toma por argumento dos enteros y retorna el máximo común divisor de ambos
  """
  while a != b:
    if a > b:
      a = a - b
    else:
      b = b - a
  return a

# Ejecución del algoritmo:

a = 48
b = 18
gcd = EuclidianAlgorithm(a, b)
print(f"El máximo común divisor de {a} y {b} es: {gcd}")
```

 El objetivo de este algoritmo es simular el cambio continuo de una variable, actualizando el valor de la posición en el paso siguiente $x(k+1)$ a partir del valor de la posición en el paso actual $(x(k))$ y la forma en que cambia $d(k)$ durante $n$ pasos. 
 
 Este es el tipo de algoritmos usados en Inteligencia Artificial para guiar los procesos de aprendizaje de las máquinas. En este ejemplo, tenemos un proceso iterativo de $n$  pasos que actualiza un arreglo $x$. El algoritmo recibe como parámetros de entrada una condición inicial $x[0]$ y el número de pasos $n$, para retornar el arreglo final $x$, que muestra los sucesivos cambios de la variable.

**2.1.** Implemente el algoritmo anterior como una función llamada ***trajectory*** que recibe como parámetros una condición inicial $x_0$ de tipo *float* y un número de pasos ***n*** de tipo *int* y realice una prueba de ejemplo.  

```python
#Ejecución: implemente y ejecute su algoritmo aquí

import numpy as np

def trajectory(x: float, n: int) -> list:
    d = []
    x = [x]
    for i in range(n):
        d.append(float(np.cos(x[i])+(10/3)*np.cos((10/3)*x[i])))
        x.append(float(x[i]-(0.1*d[i])))
    return x

print(trajectory(0.1, 10))
```
